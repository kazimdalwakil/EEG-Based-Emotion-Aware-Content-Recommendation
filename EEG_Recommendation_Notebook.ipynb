Passing through functions to make features

```python
#1-3
def mean(x):
    return (np.mean(x))
def std(x):
    return np.std(x)
def ptp(x):
    return np.ptp(x)

#4-6
def var(x):
    return np.var(x)
def minim(x):
    return np.min(x)
def maxim(x):
    return np.max(x)

#7-10
def rms(x):
    return np.sqrt(np.mean(x**2))
def mode(x):
    return st.mode(x)[0]
def skewness(x):
    return st.skew(x)
def kurtosis(x):
    return st.kurtosis(x)

# def concatenate_features(x):
#     return np.concatenate (
#         (mean(x),std(x),ptp(x), var(x),minim(x),
#         maxim(x),rms(x), mode(x),skewness(x),
#         kurtosis(x)),axis=None)

def concatenate_features(x):
    return np.concatenate (
        (mean(x),std(x),ptp(x), var(x),minim(x),
        maxim(x),rms(x)),axis=None)
```

## AHP ranking

```python
# Importing necessary packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#Importing data
ahp_df = pd.read_csv("/content/drive/MyDrive/Day 1 by Al-Wakil/AHP_PairWise_Matrix.csv")
ahp_df.set_index('Unnamed: 0', inplace=True)
ahp_df = pd.DataFrame(ahp_df)
ahp_df.head()
```

```python
def ahp_algorithm(ahp_df):
  #calculating sum of all columns
  all_columns_sum = []
  for columns in ahp_df.columns:
    sum = np.sum(np.array(ahp_df[columns]))
    all_columns_sum.append(sum)
  print("all_columns_sum: ", all_columns_sum)

  #dividing each column with the sum, Normalizing the values
  idx = 0
  normalizing_list = []

  for column in ahp_df.columns:
    normalizing_list.append(np.array(ahp_df[column])/all_columns_sum[idx])
    idx+=1
  normalizing_list = np.array(normalizing_list)
  normalizing_list = normalizing_list.T #Transposing the list
  normalizing_list_df = pd.DataFrame(normalizing_list)
  print("\nNormalized Matrix: ")
  print(normalizing_list_df)

  #Criteria Weight -> Avg. of each row of normalizing_list_df
  criteria_weight = []
  for i in range(3):
    criteria_weight.append(np.average(np.array(normalizing_list_df.loc[i])))

  print("\ncriteria_weight: ",criteria_weight) # the weights we are looking for


  # now we will check if the importance we have assigned (pair-wise matrix) is consistent or not
  # Consistency Matrix -> criteria_weight * each column values of original pair-wise matrix
  consistency_list = []
  idx = 0
  for column in ahp_df.columns:
    consistency_list.append(np.array(ahp_df[column]) * criteria_weight[idx])
    idx+=1
  consistency_list = np.array(consistency_list)
  consistency_list = consistency_list.T #Transposing the list
  consistency_list_df = pd.DataFrame(consistency_list)
  print("\nConsistency_list_df: ")
  print(consistency_list_df)

  # Sum of the rows of consistency_list_df, we will call it Weighted Sum Values(WSV)
  weighted_sum_values = []
  for i in range(3):
    weighted_sum_values.append(np.sum(np.array(consistency_list_df.loc[i])))

  print("\nweighted_sum_values: ",weighted_sum_values)

  # finding out the ratio -> WSV/CR, we will find a list of values, those values are called lambda values
  lambda_vals_ratios = []
  for i in range(len(weighted_sum_values)):
    lambda_vals_ratios.append(weighted_sum_values[i] / criteria_weight[i])

  print("\nlambda_vals_ratios: ", lambda_vals_ratios)

  #finding the Lambda_max -> Lambda_max = Average of 'lambda_vals_ratios'
  lambda_max = np.average(np.array(lambda_vals_ratios))
  print("\nlambda_max: ", lambda_max)

  #finding Consistency Index -> C.I = (Lambda_max - n)/(n-1) where n = Number of criteria (in our case it's 3)
  number_of_criteria = 3
  consistency_index = (lambda_max - number_of_criteria)/(number_of_criteria - 1)
  print("\nconsistency_index: ", consistency_index)

  #finding Consistency ratio -> C.R = C.I / Random Index (In our case, Random Index for 3x3 matrix is 0.58)
  random_index = 0.58
  consistency_ratio = consistency_index/random_index
  print("\nconsistency_ratio: ", consistency_ratio)

  # if consistency_ratio < 0.1 then the importance that we have assigned is consistent and valid
  # else: we had to change the importance vlaues (the first pair-wise matrix)
  # if consistent, we can use the criteria_weights
  print("\nConsistency Checking: ")
  if consistency_ratio < 0.1:
    print("Valid")
    return criteria_weight
  else:
    print("Not Consistend, change the importance values")
    return ["No weights"]
```
